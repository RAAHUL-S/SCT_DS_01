**Extraction of Top-10 countries with respect to Total Population**

```python
total_population_data=df[df["Series Code"]=="SP.POP.TOTL"]
total_population_sorted=total_population_data.sort_values(by="2022", ascending=False)
total_top_ten_countries=total_population_sorted.head(10)
print("Top Ten Countries of Total Population are: \n")
print(total_top_ten_countries[["Country Name"]] )
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(2, 2, 1)
sns.barplot(x="2022", y="Country Name", data=total_top_ten_countries, palette="Blues")
plt.title("Top Ten Countries of Total Population (2022)", fontsize=14)
plt.xlabel("Total Population", fontsize=12)
plt.ylabel("Country", fontsize=12)

plt.subplot(2, 2, 2)
sns.barplot(x="2016", y="Country Name", data=total_top_ten_countries, palette="Greens")
plt.title("Top Ten Countries of Total Population (2016)", fontsize=14)
plt.xlabel("Total Population", fontsize=12)
plt.ylabel("Country", fontsize=12)

plt.tight_layout()
plt.show()
```

**Top 10 Countries of Total Population in years 2010 and 2001 respectively**

```python
plt.figure(figsize=(15, 6))
plt.subplot(2, 2, 1)
sns.barplot(x="2010", y="Country Name", data=total_top_ten_countries, palette="Reds")
plt.title("Top Ten Countries of Total Population (2010)", fontsize=14)
plt.xlabel("Total Population", fontsize=12)
plt.ylabel("Country", fontsize=12)

plt.subplot(2, 2, 2)
sns.barplot(x="2001", y="Country Name", data=total_top_ten_countries, palette="Purples")
plt.title("Top Ten Countries of Total Population (2001)", fontsize=14)
plt.xlabel("Total Population", fontsize=12)
plt.ylabel("Country", fontsize=12)

plt.tight_layout()
plt.show()
```

**Top-10 countries with Highest Male and Female Population in 2022**

```python
female_population_data=df[df["Series Code"]=="SP.POP.TOTL.FE.IN"]
male_population_data=df[df["Series Code"]=="SP.POP.TOTL.MA.IN"]
male_population_sorted=male_population_data.sort_values(by="2022", ascending=False)
female_population_sorted=female_population_data.sort_values(by="2022", ascending=False)

female_top_ten_countries=female_population_sorted.head(10)
print("Top Ten Countries of Female Population: \n")
print(female_top_ten_countries[["Country Name"]])
male_top_ten_countries=male_population_sorted.head(10)
print("Top Ten Countries of Male Population: \n")
print(male_top_ten_countries[["Country Name"]])
female_population_sorted=female_population_data.sort_values(by="2022", ascending=False)
plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Name", data=male_top_ten_countries, palette="viridis")
plt.title("Top Ten Countries with Highest Male Population (2022)",size=10)
plt.xlabel("Male Population",size=10)
plt.ylabel("Country",size=10)
plt.show()

plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2022", y="Country Name", data=female_top_ten_countries, palette="viridis")
plt.title("Top Ten Countries with Highest Female Population (2022)",size=10)
plt.xlabel("Female Population",size=10)
plt.ylabel("Country",size=10)
plt.show()
```

**Top-10 countries with Lowest Male and Female Population in 2022**

```python
male_lowest_ten_countries=male_population_sorted.tail(10)
print("Top Ten Countries of Female Population: \n")
print(male_lowest_ten_countries[["Country Name"]])

female_lowest_ten_countries=female_population_sorted.tail(10)
print("Top Ten Countries of Male Population: \n")
print(female_lowest_ten_countries[["Country Name"]])

plt.figure(figsize=(15, 6))
plt.subplot(2,2,1)
sns.barplot(x="2022", y="Country Name", data=male_lowest_ten_countries, palette="PiYG")
plt.title("Top Ten Countries with Lowest Male Population (2022)",size=10)
plt.xlabel("Male Population",size=10)
plt.ylabel("Country",size=10)
plt.show()

plt.figure(figsize=(15, 6))
plt.subplot(2,2,2)
sns.barplot(x="2022", y="Country Name", data=female_lowest_ten_countries, palette="PiYG")
plt.title("Top Ten Countries with Lowest Female Population (2022)",size=10)
plt.xlabel("Female Population",size=10)
plt.ylabel("Country",size=10)
plt.show()
```

**Stacked Bar Plot**

**Top-10 countries with Male and Female Populations (2022)**

```python
merged_data=pd.merge(male_population_data, female_population_data, on="Country Name", suffixes=("_Male", "_Female"))
merged_data=merged_data[["Country Name", "2022_Male", "2022_Female"]]
merged_data=merged_data.sort_values(by="2022_Male", ascending=False)
```

```python
merged_data
```

```python
merged_data["Total Population"]=merged_data["2022_Male"]+merged_data["2022_Female"]
merged_data=merged_data.sort_values(by="Total Population", ascending=False)
```

```python
merged_data.head()
```

```python
sorted_data=merged_data.sort_values(by="Total Population", ascending=False)
top_10_countries=sorted_data.head(10)
# Create the Stacked Bar Plot
plt.figure(figsize=(12, 6))
sns.barplot(x="Country Name", y="2022_Female", data=top_10_countries, color="yellow", label="Female Population")
sns.barplot(x="Country Name", y="2022_Male", data=top_10_countries, bottom=top_10_countries["2022_Female"], color="blue", label="Male Population")
plt.title("Top 10 Countries with Male and Female Populations (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```

**Bottom 10 countries with Male and Female Population (2022)**

```python
bottom_10_countries=sorted_data.tail(10)

plt.figure(figsize=(12, 6))
sns.barplot(x="Country Name", y="2022_Female", data=bottom_10_countries, color="orange", label="Female Population")
sns.barplot(x="Country Name", y="2022_Male", data=bottom_10_countries, bottom=bottom_10_countries["2022_Female"], color="green", label="Male Population")
plt.title("Bottom 10 Countries with Male and Female Populations (2022)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```
